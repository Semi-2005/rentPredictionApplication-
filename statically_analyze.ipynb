```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('processed_data.csv')
```

```python
df['city'] = df['city'].astype('category')
df['district'] = df['district'].astype('category')
df['neighborhood'] = df['neighborhood'].astype('category')
df['room'] = df['room'].astype('int')
df['living_room'] = df['living_room'].astype('int')
df['area_mSqr'] = df['area_mSqr'].astype('int')
df['building_age'] = df['building_age'].astype('int')
df['floor'] = df['floor'].astype('float')
df['list-view-price'] = df['list-view-price'].astype('float')
```

```python
print(df.info())
```

##### Numerik degiskenlerin minimum,maximum ve ceyreklik degerlerinin bulunmasi

```python
columns = df.select_dtypes(include=[np.number]).columns ##sadece numerik olan stunlari ceviricek
min_values = []
max_values = []
for column in columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    min_value = Q1 - 1.5 * IQR
    max_value = Q3 + 1.5 * IQR
    min_values.append(min_value)
    max_values.append(max_value)
    print(f'Column: {column}, Min: {min_value}, Max: {max_value}')
```

Aykiri degerlerin temizlenmesi

```python
   for i, column in enumerate(columns):
       df = df[(df[column] >= min_values[i]) & (df[column] <= max_values[i])]

   print(df.info())
```

```python
df.to_csv('data_cleaned.csv', index=False)
```
